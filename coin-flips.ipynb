```python
import random 
import math 
import statistics 
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def coin(n):
    flip = random.randint(0,1)
    return flip
```

```python
def how_many_flips(n=2**16):  
    heads_count = 0
    tails_count = 0
    total_flips = 0
    data = []
    for x in range(n): 
        if coin(1) == True:
            heads_count += 1
            total_flips += 1
        else:
            tails_count += 1
            total_flips += 1
        if math.log2(x+1) % 1 == 0:
            data.append((heads_count,tails_count))
    if sum(data[len(data)-1]) != total_flips:
        data.append((heads_count, tails_count))
    return data

coin_flips = how_many_flips()
```

```python
print(coin_flips)    
```

```python
plt.plot([x-y for x,y in coin_flips])
plt.xlabel('Number of Flips Per Test Run')
plt.ylabel('Number of Tails Subtracted From the Number of Heads')
plt.show()
```

```python
plt.scatter([x+y for x,y in coin_flips], [x-y for x,y in coin_flips])
plt.xlabel('Number of Flips Per Test Run')
plt.ylabel('Number of Tails Subtracted From the Number of Heads')
plt.xscale('log')
plt.show()
```

```python
plt.plot([x/y for x,y in coin_flips])
plt.xlabel('Number of Flips Per Test Run')
plt.ylabel('Ratio of Heads to Tails')
plt.show()
```

```python
plt.scatter([x+y for x,y in coin_flips], [x/y for x,y in coin_flips])
plt.xlabel('Number of Flips Per Test Run')
plt.ylabel('Ratio of Heads to Tails')
plt.xscale('log')
plt.show()
```

```python
plt.plot([x/(x+y) for x,y in coin_flips])
plt.xlabel('Number of Flips Per Test Run')
plt.ylabel('Ratio of Heads to Total Coin Flips')
plt.show()
```

```python
plt.scatter([x+y for x,y in coin_flips], [x/(x+y) for x,y in coin_flips])
plt.xlabel('Number of Flips Per Test Run')
plt.ylabel('Ratio of Heads to Total Coin Flips')
plt.xscale('log')
plt.show()
```

The above graphs show that as the number of coin flips increases the ratio of heads to tails moves towards a 1 to 1 ratio and the percentage of either heads or tails flipped moves towards 50%. 

## Mean and Standard Deviation

```python
def twenty_times():
    return [how_many_flips() for x in range(20)]
```

```python
twenty_times = twenty_times()
print(twenty_times)
```

```python
def ratio(r):
    return [x/(y+x) for x,y in r]

tr = [ratio(n) for n in twenty_times]

print(tr)
```

```python
plt.scatter([x+y for x,y in coin_flips], [statistics.mean(x) for x in list(zip(*tr))])
plt.xlabel('Number of Flips Per Test Run')
plt.ylabel('Mean of Ratio of Heads')
plt.xscale('log')
plt.show()
```

```python
plt.scatter([x+y for x,y in coin_flips], [statistics.stdev(x) for x in list(zip(*tr))])
plt.xlabel('Number of Flips Per Test Run')
plt.ylabel('Standard Deviations')
plt.xscale('log')
plt.show()
```

The above graphs for mean and standard deviation show that as the number of times the coin is flipped increases the data becomes more consistent and moves towards a 1 to 1 ratio. This increasing consistency is represented by the mean approaching .50 and the standard deviation graph moving towards zero as the outliers become a smaller percent of the total data.

# 100,000 Runs

```python
def flip100():
    num_heads = 0
    for _ in range(100):
        if coin(1) == True:
            num_heads += 1
    return num_heads/100

print(flip100())
```

```python
def many_times100():
    return [flip100() for x in range(100000)]
many_times100 = many_times100()
print(many_times100)
```

```python
plt.hist(many_times100)
plt.show()
```

```python
def flip1000():
    num_heads = 0
    for _ in range(1000):
        if coin(1) == True:
            num_heads += 1
    return num_heads/1000

print(flip1000())
```

```python
def many_times1000():
    return [flip1000() for x in range(100000)]
many_times1000 = many_times1000()
```

```python
print(many_times1000)
```

```python
plt.hist(many_times1000)
plt.show()
```

```python
plt.boxplot([many_times100, many_times1000])
plt.show()
```

The above graphs show that as the number of times the coin is flipped increases the more consistent the findings become and therefore the tighter the above graphs become. This can be seen by the smaller spread of data seen in the above graphs for 100,000 trials of 1000 coin flips as compared to the graphs for 100,000 trials of 100 coin flips.
