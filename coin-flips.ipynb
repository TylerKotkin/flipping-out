```python
import random 
import math 
import statistics 
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def coin(n):
    flip = random.randint(0,1)
    return flip
```

```python
def coin_flips(n=2**16):  
    heads_count = 0
    tails_count = 0
    total_flips = 0
    data = []
    for x in range(n): 
        if coin(1) == True:
            heads_count += 1
            total_flips += 1
        else:
            tails_count += 1
            total_flips += 1
        if math.log2(x+1) % 1 == 0:
            data.append((heads_count,tails_count))
    if sum(data[len(data)-1]) != total_flips:
        data.append((heads_count, tails_count))
    return data
```

```python
print(coin_flips())    
```

```python
plt.plot([x-y for x,y in coin_flips()])
plt.xlabel('Number of Flips Per Test Run')
plt.ylabel('Number of Tails Subtracted From the Number of Heads')
plt.show()
```

```python
plt.scatter([x+y for x,y in coin_flips()], [x-y for x,y in coin_flips()])
plt.xlabel('Number of Flips Per Test Run')
plt.ylabel('Number of Tails Subtracted From the Number of Heads')
plt.xscale('log')
plt.show()
```

```python
plt.plot([x/y for x,y in coin_flips()])
plt.xlabel('Number of Flips Per Test Run')
plt.ylabel('Ratio of Heads to Tails')
plt.show()
```

```python
plt.scatter([x+y for x,y in coin_flips()], [x/y for x,y in coin_flips()])
plt.xlabel('Number of Flips Per Test Run')
plt.ylabel('Ratio of Heads to Tails')
plt.xscale('log')
plt.show()
```

```python
plt.plot([x/(x+y) for x,y in coin_flips()])
plt.xlabel('Number of Flips Per Test Run')
plt.ylabel('Ratio of Heads to Total Coin Flips')
plt.show()
```

```python
plt.scatter([x+y for x,y in coin_flips()], [x/(x+y) for x,y in coin_flips()])
plt.xlabel('Number of Flips Per Test Run')
plt.ylabel('Ratio of Heads to Total Coin Flips')
plt.xscale('log')
plt.show()
```
